**Importing Dependencies**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

**Load the dataset**

```python
file_path = '/content/drive/MyDrive/endpoint_security_logs.csv'
data = pd.read_csv(file_path)
```

**Quick look at data**

```python
data.head()
```

```python
import plotly.graph_objects as go
subset_start, subset_end = 0, 200
subset_data = data.iloc[subset_start:subset_end]
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=subset_data['Timestamp'],
    y=subset_data['Severity'],
    name='Severity Level',
    mode='lines+markers',
    marker=dict(color='blue')
))
fig.update_layout(
    showlegend=True,
    title='Severity Level Visualization (Subset of Dataset)',
    xaxis_title='Timestamp',
    yaxis_title='Severity Level',
    template='plotly_white'
)

fig.show()
```

```python
print("Dataset Head:\n", data.head())
print("\nDataset Info:\n")
data.info()
print("\nDataset Description:\n", data.describe())
```

**Convert Timestamp to datetime format**

```python
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
data.sort_values(by='Timestamp', inplace=True)
```

**Encode categorical columns**

```python
categorical_columns = ['Device_ID', 'User_ID', 'Event_Type', 'Severity', 'Source_IP', 'Action_Taken', 'Description']
encoders = {}
for col in categorical_columns:
    encoders[col] = LabelEncoder()
    data[col] = encoders[col].fit_transform(data[col])
```

**Normalize the data**

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data.drop(columns=['Timestamp']))
```

**Define anomalies as rows with Severity marked**

```python
data['Anomaly'] = data['Severity'].apply(lambda x: 1 if x > 0 else 0)
```

**Prepare data for LSTM**

```python
sequence_length = 50
X, y = [], []
for i in range(sequence_length, len(scaled_data)):
    X.append(scaled_data[i-sequence_length:i])
    y.append(data['Anomaly'].iloc[i])
X, y = np.array(X), np.array(y)
```

**Train the LSTM Model**

```python
model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(X.shape[1], X.shape[2])),
    Dropout(0.2),
    LSTM(32, return_sequences=False),
    Dropout(0.2),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=20, batch_size=32, validation_split=0.2)
```

**Predict anomalies**

```python
predictions = model.predict(X)
anomaly_indices = np.where(predictions > 0.5)[0]
```

**Plot anomalies**

```python
plt.figure(figsize=(14, 6))
plt.plot(range(len(data['Severity'])), data['Severity'], label='Severity Level')
plt.scatter(anomaly_indices, [data['Severity'].iloc[i] for i in anomaly_indices], color='red', label='Anomalies', marker='x')
plt.title('Anomaly Detection in Logs (Full Dataset)')
plt.xlabel('Index')
plt.ylabel('Severity Level')
plt.legend()
plt.show()
```

**Clearer visualization**

```python
subset_start, subset_end = 0, 200
subset_data = data.iloc[subset_start:subset_end]
subset_predictions = predictions[subset_start:subset_end]
subset_anomaly_indices = np.where(subset_predictions > 0.5)[0]

plt.figure(figsize=(14, 6))
plt.plot(range(subset_start, subset_end), subset_data['Severity'], label='Severity Level')
plt.scatter(
    subset_anomaly_indices + subset_start,
    [subset_data['Severity'].iloc[i] for i in subset_anomaly_indices],
    color='red', label='Anomalies', marker='x')
plt.title('Anomaly Detection in Logs (Subset)')
plt.xlabel('Index')
plt.ylabel('Severity Level')
plt.legend()
plt.show()
```
